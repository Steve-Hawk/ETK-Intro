# Finite differencing
## Basic concepts and Formula
Thinking in a simple mind, we can approximate a function with polynomial function using Lagrange interpolation methods, then use its derivates to approximate original results

In general, we can use the the Taylor expansion theorem to express the grid point value:

$$ f _ { i - 1 } = f _ { i } - \left. \frac { \partial f } { \partial x } \right| _ { x = x _ { i } } \Delta x + \frac { 1 } { 2 } \left. \frac { \partial ^ { 2 } f } { \partial x ^ { 2 } } \right| _ { x = x _ { i } } \Delta x ^ { 2 } - \frac { 1 } { 6 } \left. \frac { \partial ^ { 3 } f } { \partial x ^ { 3 } } \right| _ { x = x _ { i } } \Delta x ^ { 3 } + \mathcal { O } \left( \Delta x ^ { 4 } \right) \tag{1.1}$$

$$ f _ { i + 1 } = f _ { i } + \left. \frac { \partial f } { \partial x } \right| _ { x = x _ { i } } \Delta x + \frac { 1 } { 2 } \left. \frac { \partial ^ { 2 } f } { \partial x ^ { 2 } } \right| _ { x = x _ { i } } \Delta x ^ { 2 } + \frac { 1 } { 6 } \left. \frac { \partial ^ { 3 } f } { \partial x ^ { 3 } } \right| _ { x = x _ { i } } \Delta x ^ { 3 } + \mathcal { O } \left( \Delta x ^ { 4 } \right) \tag{1.2}$$

As a results:

we can express the first and second derivitave based on the grid values:
$$ \left. \frac { \partial f } { \partial x } \right| _ { x = x _ { i } } = \frac { 1 } { \Delta x } \left( f _ { i + 1 } - f _ { i } \right) + \mathscr { O } ( \Delta x ) \tag{1.3}$$
$$ \left. \frac { \partial ^ { 2 } f } { \partial x ^ { 2 } } \right| _ { x = x _ { i } } = \frac { 1 } { ( \Delta x ) ^ { 2 } } \left( f _ { i - 1 } - 2 f _ { i } + f _ { i + 1 } \right) + \mathcal { O } \left( \Delta x ^ { 2 } \right) \tag{1.4}$$

# Numerical PDE Solution
## Simple Concepts
as the example shows:
$$ \frac { \partial u } { \partial t } = D \frac { \partial ^ { 2 } u } { \partial x ^ { 2 } }\tag{2.1}$$ 
- Initial Conditions(IC) 
    $$ u ( x , t = 0 ) = u _ { 0 }\tag{2.1.1}$$
- Boundary Conditions(BC)
    - Dirichlet(the dependent spatial variable is specified)
    $$ u \left( x = x _ { 0 } , t \right) = u _ { b }\tag{2.1.2}$$
    - Neumann(the spatial derivative is specified)
    $$\frac { \partial u \left( x = x _ { f } , t \right) } { \partial x } = 0\tag{2.1.3}$$
    - Robin Conditions(Combined style)
    
An analytical solution is generally unavailable for realistic PDE problems in science and engineering, the numerical solution is an approximation to the analytical solution, and our expectation is that it represents the analytical solution with good accuracy

## Elements of MOL
To solve the simplest partial derivative equation:
$$u _ { t } + v u _ { x } = 0\tag{2.2.1}$$

One strategy is to construct a equivalent ODE, then make use of mature ODE numerical solutions
- to replace the spatial derivative with the finite differencing 
$$u _ { x } \approx \frac { u _ { i } - u _ { i - 1 } } { \Delta x }\tag{2.2.2}$$
- thus we replace the original equation with the below:
$$\frac { d u _ { i } } { d t } = - v \frac { u _ { i } - u _ { i - 1 } } { \Delta x } , 1 \leq i \leq M\tag{2.2.3}$$
- the boundary conditions:
$$u \left( x _ { i } , t = 0 \right) = f \left( x _ { i } \right) , 1 \leq i \leq M\tag{2.2.3}$$

### Key Point:
- the above procedure can be treated as a time slicing of the PDE
- time slicing gives a series of ODE equations
We reduce all the above equations as:

$$\frac { u _ { i } ^ { n + 1 } - u _ { i } ^ { n } } { \Delta t } = - v \frac { u _ { i } ^ { n } - u _ { i - 1 } ^ { n } } { \Delta x }\tag{2.3}$$

## Combine MOL with Runge-Kutta Method:
When we need to obtain more accurate approximate, we can make use of Runge-Kutta method to iteratively obtain the next time-domain values.
### Basic of Runge-Kutta Method
In short, using Euler-forward method to predict the next-grid value, we use the slope at the current point.However, we can combine the interval point with the current point with superposition to achieve at leat two order approximate.

### Initial condition setup
$$\phi ( 0 , x ) = \exp \left( - 20 x ^ { 2 } \right) , \quad \partial _ { t } \phi ( 0 , x ) = 0\tag{4.1}$$

which means:
$$\partial _ { x } \phi ( 0 , x ) = - 40 x \exp \left( - 20 x ^ { 2 } \right)$$

we express the grid value dependent on time and spatial domain as column vector

$$\frac { \partial } { \partial t } \left( \begin{array} { c } { \phi } \\ { \phi _ { t } } \\ { \phi _ { x } } \end{array} \right) = \left( \begin{array} { c } { \phi_{t} } \\ { 0 } \\ { 0} \end{array} \right) + \frac { \partial } { \partial x } \left( \begin{array} { c } { 0 } \\ { \phi _ { x } } \\ { \phi _ { t } } \end{array} \right)\tag{4.2}$$

### Construct the grid distribution

```python
def grid(Npoints):
    '''
    use cell-centered grid:
    left boundary lies between first and second grid point
    right boundary lies between last and second-to-last grid points
    '''
    x_min  = -1.0
    x_max = 1.0
    
    dx = (x_max - x_min)/Npoints
    
    x_i_min = x_min - dx/2.0
    x_i_max = x_max + dx/2.0
    
    # Arithmetic progression on grids
    return dx, np.linspace(x_i_min, x_i_max, Npoints+2)
```

### Initial conditions on the grid

```python
def initial_data(x): 
    '''
    dx, x = grid(Npoints)
    '''
    U = np.zeros_like((3,len(x)))
    U[0, :] = np.exp(-20.0 * x **2)
    U[2, :] = -40.0 * x * np.exp(-20.0 * x**2)
    
    return U
```

### partial derivative on time
slicing tips:[start_index: stop_index: step]
- start_index is the beginning of slicing
- stop_index is the final position of slicing(exclude)

```python
def RHS(U, dx):
    
    phi = U[0, :]
    phi_t = U[1, :] # the derivative with respect to time
    phi_x = U[2, :] # the derivative with respect to domain
    
    # dU_dt is matrix
    dU_dt = np.zeros_like(U)
    
    dU_dt[0, :] = phi_t
    
    dU_dt[1, 1:-1] = 1.0 / (2.0 * dx)*(phi_x[2:] - phi_x[:-2]) # 2dx interval
    
    dU_dt[2, 1:-1] = 1.0 / (2.0 * dx)*(phi_t[2:] - phi_t[:-2])
    
    return dU_dt
```

We'll enforce that the domain is periodic as a simple boundary condition. Usually this would be an outgoing wave type boundary condition, but anything that fixes the update term at the boundary is fine.

```python
def apply_boundary(dU_dt):
    '''
    periodic boundary
    '''
    dU_dt[:, 0] = dU_dt[:, -2]
    dU_dt[:, -1] = dU_dt[:, 1]
    
    return dU_dt
```

### Runge-Kutta Method

```python
def RK2_step(U, RHS, apply_boundary, dt, dx):
    '''
    2-step Runge-Kutta
    '''
    rhs = RHS(U, dx)
    rhs = apply_boundary(rhs)
    
    U_p = U + dt * rhs
    
    rhs_p = RHS(U_p , dx)
    rhs_p = apply_boundary(rhs_p)
    
    U_new = 0.5 * (U + U_p + dt * rhs_p)
    
    return U_new
```

exciting evolving!

```python
def RHS(U, dx):
    
    phi = U[0, :]
    phi_t = U[1, :]
    phi_x = U[2, :]
    
    dU_dt = np.zeros_like(U)
    
    dU_dt[0, :] = phi_t
    dU_dt[1, 1:-1] = 1.0 / (2.0*dx)*(phi_x[2:] - phi_x[:-2])
    dU_dt[2, 1:-1] = 1.0 / (2.0*dx)*(phi_t[2:] - phi_t[:-2])
    
    return dU_dt

def apply_boundaries(dU_dt):
    """
    Periodic boundaries
    """
    
    dU_dt[:, 0] = dU_dt[:, -2]
    dU_dt[:, -1] = dU_dt[:, 1]
    
    return dU_dt

def grid(Npoints):
    """
    Npoints is the number of interior points
    """
    
    dx = 2.0 / Npoints
    return dx, np.linspace(-1.0-dx/2.0, 1.0+dx/2.0, Npoints+2)

def RK2_step(U, RHS, apply_boundaries, dt, dx):
    """
    RK2 method
    """

    rhs = RHS(U, dx)
    rhs = apply_boundaries(rhs)
    Up = U + dt * rhs
    rhs_p = RHS(Up, dx)
    rhs_p = apply_boundaries(rhs_p)
    Unew = 0.5 * (U + Up + dt * rhs_p)
        
    return Unew

def initial_data(x):
    """
    Set the initial data. x are the coordinates. U (phi, phi_t, phi_x) are the variables.
    """
    
    U = np.zeros((3, len(x)))
    U[0, :] = np.exp(-20.0 * x**2)
    U[2, :] = -40.0*x*np.exp(-20.0 * x**2)
    
    return U
```

```python
Npoints = 50
dx, x = grid(Npoints)
dt = dx / 4
U0 = initial_data(x)
U = initial_data(x)
Nsteps = int(1.0 / dt)
for n in range(Nsteps):
    U = RK2_step(U, RHS, apply_boundaries, dt, dx)
```

```python
plt.figure()
plt.plot(x, U0[0, :], 'b--', label="Initial data")
plt.plot(x, U[0, :], 'k-', label=r"$t=1$")
plt.xlabel(r"$x$")
plt.ylabel(r"$\phi$")
plt.xlim(-1, 1)
plt.legend()
plt.show()
```

### Using Animation

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML

# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

line, = ax.plot([], [], lw=2)
```

```python
# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return (line,)
# animation function. This is called sequentially
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)
```
