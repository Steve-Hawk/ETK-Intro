# Numerical Methods For PDEs

We begin with the simplest example of ODE to clarify how the setup is done, including error-estimation, matching, ploting, cache-management...

$$
    u^{\prime}(t)=-a u(t), t \in(0, T], \quad u(0)=I \tag{1.1}
$$

and we use the Crank-Nicolson finite difference method to approximate the above equations.

$$
    \frac{u^{n+1}-u^{n}}{t_{n+1}-t_{n}}=-a \frac{1}{2}\left(u^{n}+u^{n+1}\right) \tag{1.2}
$$

This is a two-order approximation finite difference. The numerical solution is:

$$
    u^{n+1}=\frac{1-\frac{1}{2} a\left(t_{n+1}-t_{n}\right)}{1+\frac{1}{2} a\left(t_{n+1}-t_{n}\right)} u^{n}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

mesh = np.linspace(0,3,201)
data = np.zeros(201)
data[0] = 1
for i in range(200):
    data[i+1] = ((1-0.5*3*0.05)/(1+0.5*3*0.05))*data[i]

u_exact = 1*np.exp(-3*mesh)
```

```python
plt.plot(mesh,data,label='finite-diff')
plt.plot(mesh,u_exact,label='exact')
plt.xlabel('mesh')
plt.ylabel('data')
plt.legend()
plt.show()
```
